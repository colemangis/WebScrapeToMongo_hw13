```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.mars
collection = db.items
```

```python
# URL of page to be scraped
url = 'https://mars.nasa.gov/news/'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
```

```python
results = soup.find_all('div', class_=)
```

```python
soup.find_all('div', class_="content_title")
```

```python
# Examine the results, then determine element that contains sought info
# results are returned as an iterable list
results = soup.find_all('div', class_='content_title')
print(results)
# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return price of listing
        para = result.find('div', class_='article_teaser_body')

        # Run only if title, price, and link are available
        if (para):
            # Print results
            print('-------------')
            print(para)

    except Exception as e:
        print(e)
```

```python
# Examine the results, then determine element that contains sought info
# results are returned as an iterable list
results = soup.find_all('div', class_='list_text')

# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('a', ['target']).text
        # Identify and return price of listing
        para = result.find('div', class_='article_teaser_body').text
        # Identify and return link to listing
        #link = result.a['href']

        # Run only if title, price, and link are available
        if (title and para and link):
            # Print results
            print('-------------')
            print(title)
            print(para)
            print(link)

            # Dictionary to be inserted as a MongoDB document
            post = {
                'title': title,
                'para': para,
                'url': link
            }

            collection.insert_one(post)

    except Exception as e:
        print(e)
```

```python
# Display items in MongoDB collection
listings = db.items.find()

for listing in listings:
    print(listing)
```
